## 1. Layer và cache

- Đặt instruction **ít thay đổi** trước (COPY requirements.txt; RUN pip install) rồi mới COPY source code.
- Mỗi RUN tạo một layer: gộp lệnh khi có thể (RUN apt update && apt install -y ...).
- Dùng **.dockerignore** để không đưa file thừa (__pycache__, .git, .env) vào build context.

## 2. Image nhỏ gọn

- Base image: alpine, *-slim (vd: python:3.10-slim thay vì python:3.10).
- **Multi-stage build**: stage 1 build, stage 2 chỉ copy artifact cần chạy → image cuối nhỏ hơn.

```python
# So sánh kích thước image
!docker images --format "table {{.Repository}}	{{.Tag}}	{{.Size}}" | head -10
```

## 3. Security cơ bản

- Chạy container với **user không phải root** (USER trong Dockerfile).
- Không hardcode secret trong image; dùng env (docker compose env_file) hoặc secret management.
- Cập nhật base image định kỳ để nhận security patch.

## 5. Lệnh hữu ích

- `docker system df`: Xem dung lượng images, containers, volumes.
- `docker system prune -a`: Dọn image/container không dùng (cẩn thận).
- `docker compose logs -f service_name`: Xem log realtime.
- `docker exec -it container_name sh`: Shell vào container đang chạy.
