# Lab 4: Networks & Volumes

## Mục tiêu
- Hiểu Docker network: container gọi nhau qua tên service
- Hiểu volume: persist data, không mất khi container xóa
- Thực hành với compose: networks, volumes

## Prerequisites
- Docker Compose đang chạy


## 1. Networks

- Các container cùng **network** (trong compose: cùng tên network) có thể resolve tên service.
- VD: service `redis` có hostname `redis` trong network, service `web` có thể kết nối tới `redis:6379`.
- Driver mặc định: **bridge** (trên host).

```python
!docker network ls
```

## 2. Volumes

- **Named volume**: Docker quản lý, persist data. VD: `minio_data` trong S3_lab.
- **Bind mount**: Map thư mục host vào container. VD: `.:/app`.
- Khi container bị xóa, dữ liệu trong volume vẫn còn (trừ khi dùng `docker compose down -v`).

```python
!docker volume ls
```

## 3. Ví dụ Compose có volume

```python
%%writefile docker-compose.lab4.yml
services:
  db:
    image: redis:alpine
    volumes:
      - lab4_redis_data:/data
    networks:
      - lab4_net
volumes:
  lab4_redis_data:
networks:
  lab4_net:
    driver: bridge
```

```python
!docker compose -f docker-compose.lab4.yml up -d
!docker compose -f docker-compose.lab4.yml ps
```

```python
!docker compose -f docker-compose.lab4.yml down
print("Volume lab4_redis_data vẫn còn. Kiểm tra: docker volume ls")
```

## 4. Tóm tắt

- Network: container giao tiếp qua tên service trong cùng network.
- Volume: named volume hoặc bind mount; data persist sau khi container xóa.
