# Lab 2: Dockerfile & Build

## Mục tiêu
- Viết Dockerfile (FROM, RUN, COPY, CMD)
- Build image: docker build
- Hiểu layers và cache

## Prerequisites
- Docker đang chạy


## 1. Dockerfile cơ bản

- FROM: Image nền. WORKDIR: Thư mục làm việc.
- RUN: Chạy lệnh khi build. COPY: Đưa file vào image.
- CMD: Lệnh mặc định khi chạy container.
Mỗi instruction tạo một layer. Đặt ít thay đổi trước để tận dụng cache.

## 2. Tạo Dockerfile và build

```python
%%writefile Dockerfile.hello
FROM alpine:latest
RUN apk add --no-cache python3
WORKDIR /app
CMD ["sh", "-c", "echo Hello from custom image"]
```

```python
!docker build -f Dockerfile.hello -t lab2-hello:latest .
```

```python
!docker run --rm lab2-hello:latest
```

## 3. Tóm tắt
