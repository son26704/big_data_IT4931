## 3. Chạy container – docker run

```python
# Chạy lệnh trong container rồi thoát (alpine không có process chạy lâu)
!docker run alpine echo "Hello from container"
```

```python
# Chạy container ở chế độ nền (-d), map port 8080:80, đặt tên
!docker run -d --name lab1-nginx -p 8080:80 nginx:alpine
print("Nginx đang chạy. Mở http://localhost:8080 trên trình duyệt.")
```

## 4. docker ps, docker logs

```python
!docker ps
```

```python
!docker logs lab1-nginx --tail 5
```

## 5. docker exec – chạy lệnh trong container đang chạy

```python
!docker exec lab1-nginx cat /etc/os-release | head -2
```

## 6. Dừng và xóa container

```python
!docker stop lab1-nginx
!docker rm lab1-nginx
print("Container đã dừng và xóa.")
```

## 7. Tóm tắt

- **Image**: template read-only (vd: `nginx:alpine`).
- **Container**: instance đang chạy từ image.
- `docker run -d -p host:container --name xxx image` → chạy nền, map port, đặt tên.
- `docker ps` / `docker logs` / `docker exec` → quản lý và debug.
- `docker stop` + `docker rm` → dọn container.
