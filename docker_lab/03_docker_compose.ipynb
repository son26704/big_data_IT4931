## Mục tiêu

## Prerequisites
- Docker và Docker Compose đang chạy


## 1. Cấu trúc file docker-compose.yml

- **services**: Mỗi service = một container (image, ports, environment, volumes).
- **networks**: Các service cùng network gọi nhau qua tên service.
- **volumes**: Named volume hoặc bind mount để persist data.

## 2. Tạo file compose đơn giản

```python
%%writefile docker-compose.lab3.yml
services:
  web:
    image: nginx:alpine
    ports:
      - "8080:80"
    networks:
      - lab3_net
  redis:
    image: redis:alpine
    networks:
      - lab3_net
networks:
  lab3_net:
    driver: bridge
```

```python
!docker compose -f docker-compose.lab3.yml up -d
```

```python
!docker compose -f docker-compose.lab3.yml ps
```

## 3. Kiểm tra và dừng stack

```python
# Nginx đang listen port 8080. Có thể mở http://localhost:8080
!docker compose -f docker-compose.lab3.yml logs web --tail 3
```

```python
!docker compose -f docker-compose.lab3.yml down
```

## 4. Tóm tắt

- docker compose up -d: chạy stack. down: dừng và xóa container.
- Các lab Kafka, S3, Airflow đều dùng docker-compose.yml tương tự.
