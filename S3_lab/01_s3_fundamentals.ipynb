## 1. Object storage là gì?

**Object storage** lưu dữ liệu dưới dạng **objects** trong một không gian phẳng (flat namespace). Mỗi object gồm:
- **Data** (payload): nội dung file nhị phân
- **Metadata**: key-value (vd: Content-Type, size, custom metadata)
- **Identifier**: key hoặc object key (giống "đường dẫn" đến object)

Khác với **file system**: không có cây thư mục thật, chỉ có key dạng string (vd: `data/2024/01/file.parquet`).

Khác với **block storage**: client truy cập theo object/key, không cần quản lý block device.

## 2. So sánh Object vs File vs Block storage

| Đặc điểm | Object (S3/MinIO) | File (NFS, HDFS) | Block (EBS, SAN) |
|----------|-------------------|------------------|------------------|
| Đơn vị truy cập | Object (key) | File, path | Block |
| Namespace | Phẳng (bucket + key) | Cây thư mục | Không có namespace |
| Metadata | Rich (key-value) | Hạn chế | Rất ít |
| Scale | Rất lớn, phân tán | Lớn | Phụ thuộc thiết bị |
| Use case | Data lake, backup, static assets | Shared files, POSIX | DB, VM disk |

## 3. Amazon S3 – Khái niệm cơ bản

### Bucket
- **Bucket** là container chứa objects.
- Tên bucket global unique (trong AWS) hoặc unique trong một MinIO server.
- Có thể set policy, versioning, lifecycle cho từng bucket.

### Object & Key
- **Object** = data + metadata, được nhận diện bởi **key** (string).
- **Key** có thể có dạng giống path: `prefix/folder/file.csv` → không có "thư mục" thật, chỉ là quy ước đặt tên.
- Kích thước object: từ 0 byte đến 5TB (S3); multipart upload cho file lớn.

### Consistency
- S3: **read-after-write consistency** cho PUT mới; **eventual consistency** cho overwrite/delete (trước đây; hiện nhiều operation đã strong consistency).
- MinIO: strong consistency trong single-node/cluster.

## 4. S3 API cơ bản (sẽ dùng ở Lab 3)

- `CreateBucket` / `ListBuckets`
- `PutObject` – upload object
- `GetObject` – download object
- `ListObjectsV2` – list objects (có prefix, pagination)
- `DeleteObject` / `DeleteBucket`

Tất cả API này **tương thích** với MinIO (S3-compatible API).
